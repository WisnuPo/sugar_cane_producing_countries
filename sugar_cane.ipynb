```python
# import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.reset_option('max_columns')
```

```python
# dataset calling
sugar_data = pd.read_csv("List of Countries by Sugarcane Production.csv")
sugar_data
```

```python
# summary statistics
sugar_data.describe()
```

```python
# data types
sugar_data.dtypes
```

```python
sugar_data.head()
```

```python
# change the value of dataset
sugar_data['Production (Tons)'] = sugar_data['Production (Tons)'].str.replace('.', '')
sugar_data['Production per Person (Kg)'] = sugar_data['Production per Person (Kg)'].str.replace('.', '').str.replace(',', '.')
sugar_data['Acreage (Hectare)'] = sugar_data['Acreage (Hectare)'].str.replace('.', "")
sugar_data['Yield (Kg / Hectare)'] = sugar_data['Yield (Kg / Hectare)'].str.replace('.','').str.replace(',', '.')

sugar_data
```

```python
# see missing data
sugar_data.isnull().sum()
```

```python
# drop missing data
sugar_data.dropna(inplace=True)
```

```python
# check missing value again

sugar_data.isnull().sum()
```

```python
# delete unnecessary column
sugar_data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
sugar_data
```

```python
# change the data type
sugar_data['Production (Tons)']= sugar_data['Production (Tons)'].astype(float)
sugar_data['Production per Person (Kg)'] = sugar_data['Production per Person (Kg)'].astype(float)
sugar_data['Acreage (Hectare)'] = sugar_data['Acreage (Hectare)'].astype(float)
sugar_data['Yield (Kg / Hectare)'] = sugar_data['Yield (Kg / Hectare)'].astype(float)
```

```python
# check data type

sugar_data.dtypes
```

```python
sugar_data.head(10)
```

```python
# dataset summary

sugar_data.describe()
```

```python
# continent

sugar_data['Continent'].value_counts()
```

```python
# make bar chart continent

continent = sugar_data['Continent'].value_counts()

# defining data
x_value = continent.index
y_value = continent

# make graph

plt.figure(figsize=(9,5))
plt.bar(x_value, y_value, width=0.5)
plt.title('Sugarcane Producing Countries by Continent', fontsize=16)
plt.xlabel('Continent')
plt.ylabel('Count')
plt.xticks(rotation=0)
```

```python
top20 = sugar_data.head(20)

# make bar chart continent

continent = top20['Continent'].value_counts()

# defining data
x_value = continent.index
y_value = continent

# make graph

plt.figure(figsize=(9,5))
plt.bar(x_value, y_value, width=0.5)
plt.title('Top 20 Sugarcane Producing Countries by Total Production', fontsize=16)
plt.xlabel('Continent')
plt.ylabel('Count')
plt.xticks(rotation=25)
```

```python
# sort asc production / person
sugarData = sugar_data.sort_values(by='Production per Person (Kg)', ascending=False)
sugDat = sugarData.head(5)
```

```python
# make a pie chart
plt.figure(figsize=(9, 7))
plt.pie(sugDat['Production per Person (Kg)'], labels=sugDat['Country'], autopct='%1.1f%%')
plt.title('Sugarcane Producing Countries by Product/Person', fontsize=16)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

x = sugar_data['Acreage (Hectare)'].values
y = sugar_data['Production (Tons)'].values
x_reshape = x.reshape(-1, 1)

model = LinearRegression()

model.fit(x_reshape, y)

y_pred = model.predict(x_reshape)

plt.scatter(x_reshape, y, label='Data')
plt.plot(x_reshape, y_pred, color='red', label='Linear Regression')
plt.xlabel('Acreage')
plt.ylabel('Production')
plt.title('Linear Regression Acreage Production')
plt.legend()
plt.show
```
